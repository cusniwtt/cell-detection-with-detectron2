```python
import cv2
import numpy as np
import os
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
def display(img):
    plt.figure(figsize=(10,10))
    plt.imshow(img)
    plt.show()

def display_stack(img1, img2):
    plt.figure(figsize=(10,10))
    plt.subplot(121)
    plt.imshow(img1)
    plt.subplot(122)
    plt.imshow(img2)
    plt.show()
```

### Horizontal Shift

```python
path = 'datasets/stage1_train/1H_Nrf2_No_ADT_5_DAPI_1_0/images/1H_Nrf2_No_ADT_5_DAPI_1_0.png'
#path = 'datasets/stage1_train/1H_Nrf2_No_ADT_5_DAPI_1_0/masks/ADT_0_1_1.png'
img1 = cv2.imread(path)
img = img1
def fill(img, h, w):
    img = cv2.resize(img, (w, h), cv2.INTER_CUBIC)
    return img
        
def horizontal_shift(img, ratio=0.0):
    if ratio > 1 or ratio < 0:
        print('Value should be less than 1 and greater than 0')
        return img
    #ratio = random.uniform(-ratio, ratio)
    h, w = img.shape[:2]
    to_shift = w*ratio
    if ratio > 0:
        img = img[:, :int(w-to_shift), :]
    if ratio < 0:
        img = img[:, int(-1*to_shift):, :]
    img = fill(img, h, w)
    return img
img = horizontal_shift(img, 0.2)

display_stack(img1, img)
```

### Vertical Shift

```python
path = 'datasets/stage1_train/1H_Nrf2_No_ADT_5_DAPI_1_0/images/1H_Nrf2_No_ADT_5_DAPI_1_0.png'
#path = 'datasets/stage1_train/1H_Nrf2_No_ADT_5_DAPI_1_0/masks/ADT_0_1_1.png'
img1 = cv2.imread(path)
img = img1

def vertical_shift(img, ratio=0.0):
    if ratio > 1 or ratio < 0:
        print('Value should be less than 1 and greater than 0')
        return img
    h, w = img.shape[:2]
    to_shift = h*ratio
    if ratio > 0:
        img = img[:int(h-to_shift), :, :]
    if ratio < 0:
        img = img[int(-1*to_shift):, :, :]
    img = fill(img, h, w)
    return img

img = vertical_shift(img, 0.2)
display_stack(img1, img)
```

### Zoom

```python
path = 'datasets/stage1_train/1H_Nrf2_No_ADT_5_DAPI_1_0/images/1H_Nrf2_No_ADT_5_DAPI_1_0.png'
#path = 'datasets/stage1_train/1H_Nrf2_No_ADT_5_DAPI_1_0/masks/ADT_0_1_1.png'
img1 = cv2.imread(path)
img = img1

def zoom(img, value):
    if value > 1 or value < 0:
        print('Value for zoom should be less than 1 and greater than 0')
        return img
    h, w = img.shape[:2]
    h_taken = int(value*h)
    w_taken = int(value*w)
    #print(h_taken, w_taken)
    #h_start = random.randint(0, h-h_taken)
    #w_start = random.randint(0, w-w_taken)
    h_start = 40
    w_start = 40
    img = img[h_start:h_start+h_taken, w_start:w_start+w_taken, :]
    img = fill(img, h, w)
    return img
img = zoom(img, 0.7)

display_stack(img1, img)
```
